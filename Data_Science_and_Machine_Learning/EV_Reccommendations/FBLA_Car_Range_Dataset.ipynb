```python
data['range'] = round(data['range'] / 1.609).astype(int)
```

```python
data['range'].value_counts().sort_index()
```

```python
optimal_d = data['range'][data['range'] < 250]
sub_optimal_d = data['range'][data['range'] > 250][data['range'] < 299]
non_optimal_d = data['range'][data['range'] >= 300]

plt.hist([optimal_d, sub_optimal_d, non_optimal_d], color = ['red', 'darkorange', 'green'], bins = 50)
plt.xlabel('Range in Miles')
plt.ylabel('Number of EV Models')
plt.title('Range of All EV Models')

plt.show()
```

```python
plt.scatter(x = data['range'], y = data['range'].value_counts())
```

```python
hist, bins, patches = plt.hist(data['range'], bins=100, edgecolor='black')

# Extract bin centers and heights
bin_centers = (bins[:-1] + bins[1:]) / 2
bar_heights = hist

# Create scatterplot with marker size based on bar height
plt.scatter(bin_centers, bar_heights, marker='o', s=bar_heights * 20, alpha=0.7)
```

```python
plt.style.use('_mpl-gallery')

fig, ax = plt.subplots()

ax.stem(data['range'], y = data['range'].value_counts())

plt.show()
```
