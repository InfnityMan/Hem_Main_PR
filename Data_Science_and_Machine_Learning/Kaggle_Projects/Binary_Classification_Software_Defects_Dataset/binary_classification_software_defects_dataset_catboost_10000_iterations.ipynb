```python
train.columns = ['id', 'loc', 'v_g', 'ev_g', 'iv_g', 'n', 'v', 'l', 'd', 'i', 'e', 'b', 't', 'lo_code', 'lo_comment', 'lo_blank', 'loc_code_and_comment', 'uniq_op', 'uniq_opnd', 'total_op', 'total_opnd', 'branch_count', 'defects']
test.columns = ['id', 'loc', 'v_g', 'ev_g', 'iv_g', 'n', 'v', 'l', 'd', 'i', 'e', 'b', 't', 'lo_code', 'lo_comment', 'lo_blank', 'loc_code_and_comment', 'uniq_op', 'uniq_opnd', 'total_op', 'total_opnd', 'branch_count']
```

```python
train['defects'][train['defects'] == False] = 0
train['defects'][train['defects'] == True] = 1
train['defects'] = train['defects'].astype(int)
```

```python
y = train['defects']
del train['defects']
X = train
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```

```python
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report

catboost_classifier = CatBoostClassifier(iterations = 10000, random_seed = 1, logging_level = 'Silent')

train_pool = Pool(train_X, train_y)
catboost_classifier.fit(train_pool)

val_predictions = catboost_classifier.predict(val_X)

roc_auc = roc_auc_score(val_y, val_predictions)
print('Validation ROC AUC Score:', roc_auc_score)

accuracy = accuracy_score(val_y, val_predictions)
print('Validation Accuracy:', accuracy_score)

# Classification report
print(classification_report(val_y, val_predictions))
```

```python
pred = catboost_classifier.predict(test)
output = pd.DataFrame({'id' : test['id'], 'defects' : pred})
output.to_csv('submission.csv', index = False)
print('Your submission was successfully saved!')
```
