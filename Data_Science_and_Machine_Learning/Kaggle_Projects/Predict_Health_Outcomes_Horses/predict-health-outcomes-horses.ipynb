```python
train.fillna(0, inplace = True)
```

```python
from sklearn.model_selection import train_test_split

y = train['outcome']
del train['outcome']
X = train
X = pd.get_dummies(train)

train_X, val_X, train_y, val_y = train_test_split(X, y)
```

```python
test.isnull().value_counts()
```

```python
test.fillna(0, inplace = True)
```

```python
test = pd.get_dummies(test)
```

```python
train_X['pain_moderate'] = 0
val_X['pain_moderate'] = 0
```

```python
test['nasogastric_reflux_slight'] = 0
test['pain_slight'] = 0
test['peristalsis_distend_small'] = 0
test['rectal_exam_feces_serosanguious'] = 0
test['pain_moderate'] = 0
test = test[train_X.columns]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dtc_model = DecisionTreeClassifier(random_state = 1)
dtc_model.fit(train_X, train_y)

dtc_pred = dtc_model.predict(val_X)
```

```python
pred = dtc_model.predict(test)
```

```python
submission = pd.DataFrame({'id' : test['id'], 'label': pred})
submission.to_csv('submission.csv', index = False)
```
